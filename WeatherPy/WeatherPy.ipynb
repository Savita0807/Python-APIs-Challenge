```python
# Import Dependencies
import requests
import os
import json
import pandas as pd
import numpy as np
import time 
import matplotlib.pyplot as plt
from pprint import pprint
from api_keys import weather_api_key
from citipy import citipy
from scipy.stats import linregress
```

## Generate Cities List

```python
# Use Citipy to get the random list of cities

lats = np.random.uniform(low=-90.000, high=90.000, size=900)
langs = np.random.uniform(low=-180.000, high=180.000, size=900)
Cities_art = zip(lats,langs)
City_List =[]

for lst in Cities_art:
    city = citipy.nearest_city(lst[0], lst[1])
    City_List.append(city.city_name) 
print(len(City_List))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Base weather URL
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

# Set up variables.
City_count = 0
Set_count = 1

# set up lists to hold reponse info
city = []
lat = []
lang = []
temp_max = []
humidity = []
cloudiness = []
windspeed =[]
country = []
date = []

# Loop through the list of cities and perform a request for data on each City
for cities in City_List:
    response = requests.get(query_url + cities).json()
    
    if City_count < 50:
        City_count = City_count + 1
    else:
        Set_count = Set_count + 1  
        City_count = 1
        
    try:
        print(f"Processing Record " + str(City_count) + " Set " + str(Set_count) + "| " + response['name'])
        city.append(response['name'])
        lat.append(response['coord']['lat'])
        lang.append(response['coord']['lon'])
        temp_max.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windspeed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
    except (NameError, KeyError):
        print ('City not found. Skipping...')      
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a data frame from city name, lat, lon, max temp, humidity, cloudiness, windspees, country and date
weather_dict = {
    "City": city,
    "Lat": lat,
    "Lng": lang,
    "Max Temp": temp_max,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": windspeed,
    "Country": country,
    "Date": date
}
weather_data = pd.DataFrame(weather_dict)
weather_data_unique = weather_data.drop_duplicates(subset=['City'], keep='last')

weather_data_unique.to_csv("weather_data.csv", index=False, header=True)
weather_data_unique
```

## Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

```python
# Define function to create scatter plot for city latitude and other weather measurements
def plot_grp(y,y_label):
    axs_y = plt.scatter(Clean_city_data['Lat'],Clean_city_data[y],alpha=0.75,edgecolors='black')
    plt.title(f"City Latitude vs. {y} ({time.strftime('%m/%d/%y')})")
    plt.xlabel('Latitude') 
    plt.ylabel(y_label)
    plt.xlim(-80,100)
    plt.ylim(min(Clean_city_data[y])-20,max(Clean_city_data[y])+20)
    plt.grid()
    plt.savefig(os.path.join('Images',f'City Latitude_vs_{y}.png'))
```

## Linear Regression

```python
# Create Df for Northen Hemisphere data

North_Hemisphere = Clean_city_data[Clean_city_data['Lat'] >= 0]

# Define function to create liner regression plot for Northen Hemisphere latitude and other weather measurements

def plot_line_north(y,y_label):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(North_Hemisphere['Lat'], North_Hemisphere[y])
    regress_values = North_Hemisphere['Lat'] * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(North_Hemisphere['Lat'],North_Hemisphere[y])
    plt.plot(North_Hemisphere['Lat'],regress_values,"r-")
    plt.annotate(line_eq,(40,20),fontsize=15,color="red")
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    print(f"The r-squared is: {rvalue**2}")
    plt.savefig(os.path.join('Images',f'North Hemisphere Latitude_vs_{y}.png'))
```

```python
# Create Df for Southern Hemisphere data
South_Hemisphere = Clean_city_data[Clean_city_data['Lat'] < 0]
# South_Hemisphere

# Define function to create liner regression plot for Southern Hemisphere latitude and other weather measurements

def plot_line_south(y,y_label):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(South_Hemisphere['Lat'], South_Hemisphere[y])
    regress_values = South_Hemisphere['Lat'] * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(South_Hemisphere['Lat'],South_Hemisphere[y])
    plt.plot(South_Hemisphere['Lat'],regress_values,"r-")
    plt.annotate(line_eq,(-50,35),fontsize=15,color="red")
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    print(f"The r-squared is: {rvalue**2}")
    plt.savefig(os.path.join('Images',f'South Hemisphere Latitude_vs_{y}.png'))
```
